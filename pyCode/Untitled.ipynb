```python
import numpy as np
import pandas as pd
import re 
```

```python
appendix_filename = 'data/full_appendix_sorted.txt'
symptoms_filenames = ['data/Acute_Coronary_Syndromes_clean.txt',
                    'data/Arterial_Hypertension_clean.txt',
                     'data/Atherosclerosis_clean.txt',
                         'data/Coronary_Artery_Disease_clean.txt',
                             'data/Hypertensive_Emergencies_clean.txt']
```

```python
# Get the symptoms text
symptom_file = open(symptoms_filenames[0], 'r')
symptom_txt = symptom_file.read()
```

```python
# Split them into a vector of individual words
symptom_words = re.split('\n| ', symptom_txt) 
```

```python
# Get the appendix text
appendix_file = open(appendix_filename, 'r')
appendix_txt = appendix_file.read()
```

```python
# Create a hashmap of words in the appendix
appendix_words = re.split('\n', appendix_txt)
hm = dict.fromkeys(appendix_words , 1)

# Make sure to remove the empty string if it is present
if '' in hm: 
    del hm['']
```

```python
n = len(symptom_words)

for i in range(n): 
    if symptom_words[i] in hm: 
        print(symptom_words[i])
    if i < n - 1 and symptom_words[i] + ' ' + symptom_words[i + 1] in hm: 
        print(symptom_words[i] + ' ' + symptom_words[i + 1])
    if i < n - 2 and symptom_words[i] + ' ' + symptom_words[i + 1] + ' ' + symptom_words[i + 2] in hm: 
        print(symptom_words[i] + ' ' + symptom_words[i + 1] + ' ' + symptom_words[i + 2])
```
